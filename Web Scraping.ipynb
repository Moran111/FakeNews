# web Scraping to get more data

```python
import requests
from bs4 import BeautifulSoup
```

```python
# try to scrap a news from bbc news site 
```

```python
#specify the url
one_news_site = "http://www.bbc.com"

#Query the website and return the html to the variable 'page'
page = requests.get(one_news_site)

#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```
